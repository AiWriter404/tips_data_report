```python
# Liberary of numpy
import numpy as np
```

```python
# Numpy array
# 1D array

arr = np.array([1, 2, 3, 4, 5])
print("Numpy Array:", arr)
```

```python
# Numpy array 2D
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Numpy Array:\n", arr_2d)
```

```python
# Numpy array 3D
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("3D Numpy Array:\n", arr_3d)
```

```python
# Numpy Advanced Operations
# Array reshaping
reshaped_arr = arr_2d.reshape(3, 2)
print("Reshaped Array:\n", reshaped_arr)
# Array concatenation
concatenated_arr = np.concatenate((arr, arr))
print("Concatenated Array:", concatenated_arr)
# Array slicing
sliced_arr = arr_2d[:, 1]
print("Sliced Array:", sliced_arr)
```
